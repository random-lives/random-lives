```python
import dill
import copy
import generation
from generation import reset_to_stage, generate_narrative_plan, generate_narrative
from llm_utils import GenerationContext

with open('test_examples.pkl', 'rb') as f:
    all_people = dill.load(f)
```

```python
# List test people
for i, p in enumerate(all_people):
    loc = p.location.country if p.location else p.region
    name = (p.name or "(unnamed)")[:25]
    print(f"{i:2d}. {name:25s} | {p.birth_year_str:10s} | age {str(p.age_at_death):5s} | {loc}")
```

```python
for person in all_people[:10]:
    reset_to_stage(person, 'narrative')
    ctx = GenerationContext(model='gpt-5.2', quiet=True, show_cost=True)
    generate_narrative_plan(person, ctx)
    generate_narrative(person, ctx)
    ctx.finish()
```
